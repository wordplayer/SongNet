## Step 1: Convert .png to numpy arrays

4. Walk folder and convert png's to .npy

```python
base_path = '/datasets/home/21/321/ee228sp20ta1/G51/'
train_path = '/datasets/home/21/321/ee228sp20ta1/G51/train_specs/'
val_path = '/datasets/home/21/321/ee228sp20ta1/G51/val_specs/'
# test_path = '/datasets/home/21/321/ee228sp20ta1/G51/spectrograms/test/'
```

```python
path, dirs, files = next(os.walk(val_path))
file_count = len(files)
print(file_count)
```

```python
list_specs = np.ndarray(shape=(2763,96,216))
list_ids = []

name = 'train'
idx = 0
```

```python
for image_path in tqdm(glob.glob(train_path + "/*.png")):
    
    
    # Load and collect song ids into list of strings
    ID = os.path.basename(image_path).partition(".png")[0]

    list_ids.append(ID)

    image = Image.open(image_path)

    # Load and collect spectrograms in numpy (N,x,y)
    data = asarray(image)

    list_ids.append(data)
    specs = asarray(list_specs)
    
    idx += 1
    time.sleep(0.01)
    if idx == 15000:

        # Save out song_ids in .txt
        f = open(base_path+'{}_ids.txt'.format(name), 'w')
        for song_id in list_ids:
            f.write(song_id+'\n')
        f.close()

        # Save out spectrograms in numpy 
        save(base_path+'{}_set.npy'.format(name),specs)
```
